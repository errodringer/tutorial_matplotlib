### Primera gráfica lineal con Matplitlib

Gráfica sencilla X-Y de dos arrays

```python
x = np.linspace(0, 10, 100)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x, y)
plt.show()
```

### Gráfica lineal "completa"

Cómo darle nombre a los ejes, título de la gráfica, añadir rejilla, modificar el tamaño...

```python
plt.figure(figsize=(15, 7))

plt.plot(x, y)

plt.title('f(x) -> seno(x)', fontsize=20)
plt.xlabel('x', fontsize=20)
plt.ylabel('seno(x)', fontsize=20)

plt.xlim(0, 10)
plt.ylim(-1.1, 1.1)

plt.yticks(fontsize=15)
plt.xticks(np.arange(0, 10.5, 0.5), fontsize=15)

plt.grid()

plt.savefig('funcion_seno.png')

plt.show()
```

### Gráficas múltiples

Varias gráficas en un mismo lienzo

```python
plt.plot(x, y)
y2 = np.cos(x)
plt.plot(x, y2, '--')
plt.show()
```

### Subplots

Varias gráficas en un mismo lienzo, pero diferentes ejes

```python
plt.subplot(2, 1, 1) # (filas, columnas, número de panel)
plt.plot(x, y)

plt.subplot(2, 1, 2)
plt.plot(x, y2, color='r')

plt.show()
```

### Subplots

Con títulos, etiqueta de ejes...

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2,
                               figsize=(14, 5))

ax1.scatter(x=x, y=y, marker='o', c='r', edgecolor='b')
ax1.set_title('Scatter: x versus sen(x)')
ax1.set_xlabel('$x$')
ax1.set_ylabel('$seno(x)$')

ax2.plot(x, y2, color='r',
         label=('cos(x)'))
ax2.legend(loc='best')
ax2.set_title('Gráfico lineal x versus cos(x)')
ax2.yaxis.tick_right()
ax2.set_xlabel('$x$')
ax2.set_ylabel('$coseno(x)$')

plt.show()
```

### Gráfico de barras

```python
df_group = df.groupby(['marital-status'])['age'].count().reset_index()
df_group.columns = ['marital-status', 'number']
df_group = df_group.sort_values(by=['number'], ascending=False)
```

```python
plt.bar(df_group['marital-status'], df_group['number'], color='orange')
plt.xticks(rotation=70)
plt.xlabel('Estado civil')
plt.ylabel('Nº. de casos')
plt.title('Nº de personas por estado civil')
plt.show()
```

Gráfico de barras horizontal

```python
plt.barh(df_group['marital-status'], df_group['number'], color='orange', alpha=0.3)
plt.xticks(rotation=70)
plt.xlabel('Nº. de casos')
plt.ylabel('Estado civil')
plt.title('Nº de personas por estado civil')

plt.show()
```

Barras con diferents propiedades

```python
df_group = df.groupby(['marital-status', 'salary'])['age'].count().reset_index()
df_group.columns = ['marital-status', 'salary', 'number']
df_group = df_group.sort_values(by=['number'], ascending=False)
```

```python
df_group_menor = df_group[df_group.salary==" <=50K"]
df_group_mayor = df_group[df_group.salary==" >50K"]
plt.bar(df_group_menor['marital-status'], df_group_menor['number'], color='orange', label=' <=50K')
plt.bar(df_group_mayor['marital-status'], df_group_mayor['number'],label=' >50K')
plt.xticks(rotation=70)
plt.xlabel('Estado civil')
plt.ylabel('Nº. de casos')
plt.title('Nº de personas por estado civil')
plt.legend()
plt.show()
```

```python
tcks = df_group_menor.index
tcks_mayor = df_group_mayor.index
plt.bar(tcks-0.25, df_group_menor['number'],
        color='orange', label=' <=50K', width=0.75, align='center', alpha=0.7)
plt.bar(tcks_mayor-0.75, df_group_mayor['number'],
        label=' >50K', width=0.75, align='center', alpha=0.7)
plt.xticks(tcks, labels=df_group_menor['marital-status'], rotation=70)
plt.xlabel('Estado civil')
plt.ylabel('Nº. de casos')
plt.title('Nº de personas por estado civil')
plt.legend()
plt.show()
```

### Gráficos temporales

#### AirQuality DataSet

Información y descarga del conjunto de datos: https://archive.ics.uci.edu/ml/datasets/Air+Quality

```python
df2 = pd.read_excel('data/AirQualityUCI.xlsx')
```

Info. columnas:

- Date (DD/MM/YYYY)
- Time (HH.MM.SS)
- True hourly averaged concentration CO in mg/m^3 (reference analyzer)
- PT08.S1 (tin oxide) hourly averaged sensor response (nominally CO targeted)
- True hourly averaged overall Non Metanic HydroCarbons concentration in microg/m^3 (reference analyzer)
- True hourly averaged Benzene concentration in microg/m^3 (reference analyzer)
- PT08.S2 (titania) hourly averaged sensor response (nominally NMHC targeted)
- True hourly averaged NOx concentration in ppb (reference analyzer)
- PT08.S3 (tungsten oxide) hourly averaged sensor response (nominally NOx targeted)
- True hourly averaged NO2 concentration in microg/m^3 (reference analyzer)
- PT08.S4 (tungsten oxide) hourly averaged sensor response (nominally NO2 targeted)
- PT08.S5 (indium oxide) hourly averaged sensor response (nominally O3 targeted)
- Temperature in Â°C
- Relative Humidity (%)
- AH Absolute Humidity

```python
df2['Date_Time'] = df2['Date'].astype('str')+' '+df2['Time'].astype('str')
df2['Date_Time'] = pd.to_datetime(df2['Date_Time'], format="%Y%m%d %H:%M:%S")
```

```python
df2.head()
```

### Múltiples gráficos, diferente eje Y mismo eje X

```python
time_cols = ['T', 'RH', 'AH']
x = ['Date_Time']

fig, axs = plt.subplots(len(time_cols), 1, sharex=True, figsize=(14,10))

# Remove horizontal space between axes
fig.subplots_adjust(hspace=0)

for i, col in enumerate(time_cols):
    axs[i].plot(df2[x], df2[col])
    axs[i].set_xlabel('Fecha')
    axs[i].set_ylabel(time_cols[i])

plt.show()
```

```python
time_cols = ['T', 'RH', 'AH']
colors = ['r', 'g', 'b']
x = ['Date_Time']

fig, axs = plt.subplots(len(time_cols), 1, sharex=True, figsize=(14,10))

# Remove horizontal space between axes
fig.subplots_adjust(hspace=0)

for i, col in enumerate(time_cols):
    axs[i].plot(df2[x],
                df2[col].apply(lambda x: np.nan if x == -200 else x),
                c=colors[i])
    axs[i].set_xlabel('Fecha')
    axs[i].set_ylabel(time_cols[i])

plt.show()
```

### Dos gráficas en una

```python
fig = plt.figure(figsize=(15,7))

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.05, 0.65, 0.5, 0.3])

ax1.set_title('NOx(GT) en el tiempo')
ax2.set_title('Temperatura en el tiempo')

# hay valores "extraños entorno a -200"
ax1.plot(df2['Date_Time'], df2['NOx(GT)'], color='orange')
ax2.plot(df2['Date_Time'], df2['T'], color='skyblue')

plt.show()
```

```python
fig = plt.figure(figsize=(15,7))

ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.05, 0.65, 0.5, 0.3])

ax1.set_title('NOx(GT) en el tiempo')
ax2.set_title('Temperatura en el tiempo')

# hay valores "extraños entorno a -200"
ax1.plot(df2['Date_Time'], df2['NOx(GT)'].apply(lambda x: np.nan if x == -200 else x), color='orange')
ax2.plot(df2['Date_Time'], df2['T'].apply(lambda x: np.nan if x == -200 else x), color='skyblue')

plt.show()
```

```python
df2_redu = df2.iloc[:200,:]

x = df2_redu['Date_Time']
y = np.vstack([df2_redu['RH'].apply(lambda x: np.nan if x == -200 else x),
               df2_redu['T'].apply(lambda x: np.nan if x == -200 else x)])
```

### StackPlot

```python
labels = ['Humedad R.', 'Temperatura']
plt.figure(figsize=(15,7))
plt.stackplot(x, y, labels=labels)
plt.legend(loc='best')
plt.show()
```

### Gráfica lineal con relleno

y línea horizontal en la media

```python
plt.figure(figsize=(15,7))

plt.fill_between(df2_redu['Date_Time'], df2_redu['T'].apply(lambda x: np.nan if x == -200 else x), color="skyblue", alpha=0.4)

plt.plot(df2_redu['Date_Time'], df2_redu['T'].apply(lambda x: np.nan if x == -200 else x), color = 'Slateblue', alpha = 0.6)

media = df2_redu['T'].apply(lambda x: np.nan if x == -200 else x).mean()
plt.axhline(y=media, color='r', linestyle='-')

plt.title('Temperatura registrada', fontsize = 18, color = 'black')
plt.yticks(fontsize = 12, color = 'black')
plt.xticks(fontsize = 12, rotation = 0, color = 'black')
plt.xlabel('Año, mes, día', fontsize = 14, color = 'black')
plt.ylabel('Temperatura (ºC)', fontsize = 14, color = 'black')
# plt.savefig('area_simple.png')

plt.show()
```

### Gráfico en 3D

```python
%matplotlib notebook

fig = plt.figure(figsize=(14,8))

ax = fig.add_subplot(111, projection='3d')

x = df2['NO2(GT)'].apply(lambda x: np.nan if x == -200 else x)
y = df2['NOx(GT)'].apply(lambda x: np.nan if x == -200 else x)
z = df2['T'].apply(lambda x: np.nan if x == -200 else x)

ax.scatter(x, y, z, c='r', alpha=0.5)

ax.set_xlabel('NO2(GT)')
ax.set_ylabel('NOx(GT)')
ax.set_zlabel('T')

for angle in range(0, 360, 2):
    ax.view_init(30, angle)
    plt.draw()
    plt.pause(.001)
```
